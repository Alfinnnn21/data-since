```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('covid_19_indonesia_time_series_all.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
df.describe()
df.columns
df['Date'] = pd.to_datetime(df['Date'])
```

- Baris df.info() digunakan untuk menampilkan ringkasan singkat dari DataFrame. Ini memberikan informasi tentang tipe data indeks, tipe data kolom, jumlah nilai non-null, dan penggunaan memori.
- Baris df.describe() menghasilkan statistik deskriptif untuk kolom-kolom numerik dalam DataFrame. Ini menghitung ukuran seperti jumlah (count), rata-rata (mean), standar deviasi (standard deviation), nilai minimum (minimum), nilai maksimum (maximum), dan kuartil (25%, 50%, dan 75%). Ini membantu dalam memahami distribusi dan rentang data numerik.
- Baris df.columns mengembalikan daftar nama-nama kolom dalam DataFrame. Ini berguna untuk melihat semua nama kolom.
- Baris df['Date'] = pd.to_datetime(df['Date']) sangat penting untuk memastikan bahwa kolom 'Date' diperlakukan sebagai tipe data tanggal/waktu.

```python
import matplotlib.pyplot as plt

national = df[df['Location'] == 'Indonesia'] #data yang di input nasional
plt.figure(figsize=(12,6)) # mengatur figure dalam inci
plt.plot(national['Date'], national['Total Cases'], label='Total Kasus')
plt.plot(national['Date'], national['Total Recovered'], label='Sembuh')
plt.plot(national['Date'], national['Total Deaths'], label='Meninggal')
plt.legend()
plt.title("Perkembangan Kasus COVID-19 di Indonesia ") #judul di atas
plt.xlabel("Tanggal")
plt.ylabel("Jumlah Kasus")
plt.grid(True)
plt.show()
```

- Panggilan plt.plot() yang pertama memplot data 'Total Cases' dari DataFrame national terhadap 'Date' yang sesuai. label='Total Kasus' memberikan label pada garis ini, yang akan digunakan dalam legenda.
- Panggilan plt.plot() yang kedua melakukan hal yang sama untuk 'Total Recovered', memberi label 'Sembuh'.
- Panggilan plt.plot() yang ketiga memplot 'Total Deaths', memberi label 'Meninggal'. Dengan memplot ketiga metrik ini terhadap 'Date', kita dapat memvisualisasikan trennya dari waktu ke waktu.


Grafik ini menunjukkan pertumbuhan kumulatif kasus COVID-19 di Indonesia.

Garis merah menunjukkan total kasus.

Garis hijau menunjukkan total sembuh.

Garis hitam menunjukkan total meninggal.

  -KESIMPULAN UMUM-

Tingkat Kesembuhan Tinggi: Mayoritas pasien berhasil sembuh.

Kematian Relatif Rendah: Dibandingkan total kasus, angka kematian cukup rendah.

Perlu Waspada: Lonjakan kasus yang drastis menunjukkan pentingnya upaya pencegahan dan pengendalian wabah.

```python
import seaborn as sns

jabar = df[df['Location'] == 'Jawa Barat']

# Visualisasi kasus harian baru
plt.figure(figsize=(12,6))
sns.lineplot(data=jabar, x='Date', y='New Cases')
plt.title('Kasus Harian Baru di Jawa Barat')
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Kasus Baru')
plt.grid(True)
plt.show()
```

```python
# Data terbaru untuk semua provinsi
latest_date = df['Date'].max()
latest = df[(df['Date'] == latest_date) & (df['Location Level'] == 'Province')]

# Urutkan berdasarkan total kasus
latest = latest.sort_values(by='Total Cases', ascending=False)

plt.figure(figsize=(10,8))
sns.barplot(data=latest, y='Location', x='Total Cases', palette='Reds_r')
plt.title('Total Kasus COVID-19 per Provinsi')
plt.xlabel('Total Kasus')
plt.ylabel('Provinsi')
plt.show()
```

Warna merah tua = lonjakan kasus.

Warna putih/pucat = sedikit atau tidak ada kasus.

Sangat berguna untuk melihat tren besar secara cepat.
