# 2. Python Data Analysis (pandas)

```python
import pandas as pd
```

## 2.1. Membuat DataFrame

```python
df = pd.DataFrame({'nama' : ["a","b","c"],
                   'usia' : [17,19,18],
                   'tempat tinggal' : ["Jakarta", "Medan", "Bandung"]})
df
```

## 2.2. Membaca DataFrame

```python
df = pd.read_csv('https://raw.githubusercontent.com/reynoldms/csv_data/main/Automobile_data.csv')
```

```python
df
```

```python
## membaca n data pertama
df.head(4)
```

```python
## membaca n data terakhir
df.tail(4)
```

```python
## melihat detail kolom
df.info()
```

```python
## statistic descriptive
df.describe()
```

```python
## filter make == "audi"
df[df['make'] == 'audi']
```

```python
## filter 'fuel-type' == 'diesel' & 'make' == 'toyota'
df[(df['fuel-type']=='diesel') & (df['make']=='toyota')]
```

```python
## slicing kolom berdasarkan nama kolom
df[['symboling','make']]
```

```python
## slicing baris dan kolom berdasarkan urutan
df.iloc[2:5,1:5]
```

```python
## slicing baris dan kolom berdasarkan index
df.iloc[[1,4,18],[2,10]]
```

```python
##  Pandas Aggregation (1)
### melihat jumlah item unique per kolom
df['make'].value_counts()
```

```python
##  Pandas Aggregation (2)
### berdasarkan mean, mode, median, count
df.groupby(['make', 'fuel-type'])['length'].median()
```
